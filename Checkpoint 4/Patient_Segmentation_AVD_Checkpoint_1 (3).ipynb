## **Data Construction**

Saya akan menambahkan kolom untuk mengetahui apakah jarak kunjungan seorang pasien itu sering atau jarang dan lama atau cepat (datang laginya lama atau cepat)

```python
def seringan(Annual_Visits):
    if Annual_Visits > 5:
        return 'Sering'
    elif Annual_Visits < 6:
        return 'Jarang'
    else:
        return 'Breakeven'

df['Apakah_Sering'] = df['Annual_Visits'].apply(seringan)
```

```python
def kelamaan(Days_Since_Last_Visit):
    if Days_Since_Last_Visit > 150:
        return 'Lama'
    elif Days_Since_Last_Visit < 151:
        return 'Cepat'
    else:
        return 'Breakeven'

df['Cepat_Tidak'] = df['Days_Since_Last_Visit'].apply(kelamaan)
```
